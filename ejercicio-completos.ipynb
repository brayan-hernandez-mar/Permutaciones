# Ejercicio 10


Sólo para practicar. Se tiene la palabra MURCIÉLAGO. Si no se permiten repeticiones al usar
las letras que la forman, determina:

a) El número de palabras que puedes crear.

b) El número de palabras con cinco letras.

c) La probabilidad de que una palabra de cinco letras termine en vocal.

```python
# codigo 
# a
from math import factorial
from itertools import permutations

# Palabra original
palabra_original = "MURCIÉLAGO"

# Número de letras diferentes
num_letras = len(set(palabra_original))  # Usamos set() para contar letras únicas

# Número de palabras que se pueden crear sin repetición
num_palabras_sin_repeticion = factorial(num_letras)
print(f"a) Número total de palabras posibles: {num_palabras_sin_repeticion}")

# b
letras = set(palabra_original)
palabras_5_letras = set(permutations(letras, 5))
num_palabras_5_letras = len(palabras_5_letras)
print(f"b) Número de palabras con cinco letras: {num_palabras_5_letras}")

# c
probabilidad = 15120 / 3628800
print(f"c) Probabilidad de que una palabra de cinco letras termine en vocal: {probabilidad}")
```

# Ejercicio 11

Parque vehicular. La Ciudad de México tiene placas de automóviles con tres números seguidos

por tres letras, es decir, son placas del tipo 999-AAA. Sin embargo, no pueden empezar con 0

(cero) ni se utilizan las letras I, O, Q y Ñ.

a) ¿Cuántas placas diferentes son posibles?

b) ¿Cuántas placas diferentes terminan en vocal?

c) ¿Cuál es la probabilidad de que una placa, elegida al azar, sea impar?



```python
# Codigo 
# a
total_letras = 24 ** 3
total_numeros = 1000
total_placas = total_letras * total_numeros
print(f"a) Total de placas posibles: {total_placas}")

# b
vocales = 5
placas_terminantes_en_vocal = total_letras * vocales
print(f"b) Placas que terminan en vocal: {placas_terminantes_en_vocal}")

# c
numeros_impares = 5
probabilidad_placa_impar = numeros_impares / total_numeros
print(f"c) Probabilidad de placa impar: {probabilidad_placa_impar:.3f}")
```

# Ejercicio 12

Servicios restauranteros. Una cadena de hamburguesas anuncia que se ofrecen 512 maneras de

pedir una hamburguesa. Un cliente puede elegir cualquiera de estos ingredientes: 

cátsup, mostaza, mayonesa, aderezo de la casa, cebolla, pepinillos, jitomate, tocino y lechuga. ¿Crees que lo

que afirma la empresa sea cierto? Justifica tu respuesta.

```python
# Codigo 
import math

def calcular_combinaciones(n):
    total_combinaciones = 0
    for r in range(1, n + 1):
        total_combinaciones += math.comb(n, r)
    return total_combinaciones

ingredientes = 9
total_formas = calcular_combinaciones(ingredientes)

print(f"Total de formas posibles de pedir una hamburguesa: {total_formas}")
```

# Ejercicio 13

Cultura. Una persona tiene un libro de cada uno de los autores siguientes: Mario Vargas Llosa,

Alberto Ruy Sánchez, Gabriel García Márquez, Günther Grass, Patrick Süskind, Heinrich Böll,

Arturo Pérez-Reverte y Camilo José Cela. Como aún no los ha leído, quiere ponerlos en una

repisa cerca de su cama. De acuerdo con esto, determina:

a) ¿De cuántas formas puede acomodar los libros sobre la repisa?

b) ¿De cuántas maneras puede ordenarlos si quiere que los libros de autores latinoamericanos
estén juntos?

c) ¿De cuántas formas puede acomodarlos, si los autores latinoamericanos deben estar juntos,
así como los autores alemanes y los españoles también?

d) Determina las probabilidades correspondientes a los incisos (b) y (c).

```python
# Codigo 
# a
from math import factorial
total_ways = factorial(8)
print(f"a) Hay {total_ways} formas de acomodar los libros sobre la repisa.")

# b 
latin_american_books = 3
other_books = 5
latin_american_permutations = factorial(latin_american_books)
other_permutations = factorial(other_books)
total_ways_with_latin_together = latin_american_permutations * other_permutations
print(f"b) Hay {total_ways_with_latin_together} formas de ordenar los libros con los latinoamericanos juntos.")

# c
import math

latin_american_books = 3
german_books = 2
spanish_books = 2
other_books = 1

total_ways_with_all_together = math.factorial(latin_american_books) * (math.factorial(latin_american_books) * math.factorial(german_books) * math.factorial(spanish_books))

print(f"c) Hay {int(total_ways_with_all_together)} formas de ordenar los libros con los latinoamericanos, alemanes y españoles juntos.")

# d
prob_latin_together = total_ways_with_latin_together / total_ways
prob_all_together = total_ways_with_all_together / total_ways
print(f"d) La probabilidad de ordenar los libros con los latinoamericanos juntos es {prob_latin_together:.3f}.")
print(f"   La probabilidad de ordenar los libros con los latinoamericanos, alemanes y españoles juntos es {prob_all_together:.3f}.")
```

# Ejercicio 14

Finanzas. Se va a formar un portafolio de inversión de cuatro instrumentos financieros a partir

de cinco de renta fija y seis de renta variable. ¿De cuántas maneras podría formarse el portafolio

si:

a) No hay restricciones?

b) Debe haber dos instrumentos de renta fija y dos de renta variable?

c) No debe haber ningún instrumento de renta fija?

d) Debe haber tres instrumentos de renta fija?

e) Calcula las probabilidades para los incisos anteriores.



```python
# codigo 
# a
from math import factorial, comb
num_renta_fija = 5
num_renta_variable = 6
total_instrumentos = num_renta_fija + num_renta_variable
num_seleccion = 4

combinaciones_posibles = factorial(total_instrumentos) // (factorial(num_seleccion) * factorial(total_instrumentos - num_seleccion))

print(f"Número de formas de formar el portafolio: {combinaciones_posibles}")

# b
num_seleccion_rf = 2
num_seleccion_rv = 2

combinaciones_rf = comb(num_renta_fija, num_seleccion_rf)
combinaciones_rv = comb(num_renta_variable, num_seleccion_rv)
combinaciones_totales = combinaciones_rf * combinaciones_rv
print(f"Número de formas de formar el portafolio con 2 de renta fija y 2 de renta variable: {combinaciones_totales}")

# c
combinaciones_rv = comb(num_renta_variable, num_seleccion_rv)
print(f"Número de formas de formar el portafolio sin renta fija: {combinaciones_rv}")
```

# Ejercicio 15

Futbol. Tres parejas (tres hombres y tres mujeres) compran boletos para ir a un partido de los

Pumas. Cuando llegan al estadio de Ciudad Universitaria les surge la duda de cómo sentarse en

la fila. ¿De cuántas maneras pueden hacerlo si:

a) No hay restricciones?

b) Los hombres deben sentarse juntos?

c) Deben sentarse por parejas?

d) ¿Cuál es la probabilidad de cada uno de los eventos descritos en los incisos anteriores?

```python
#codigo 
# a 
from math import factorial

num_personas = 6
cant_maneras_a = factorial(num_personas)
print("Sin restricciones: ", cant_maneras_a)

# b
num_mujeres = 3
num_hombres = 3
cant_maneras_b = factorial(num_mujeres) * factorial(num_hombres)
print("Hombres juntos: ",cant_maneras_b )

# c
num_pares = 3
cant_maneras_c = factorial(num_pares) * 2**num_pares
print("Por pareja: ", cant_maneras_c)

# d
total_p = cant_maneras_a
p_a = cant_maneras_a / total_p 
p_b = cant_maneras_b / total_p 
p_c = cant_maneras_c / total_p 
print("Probabilidad: ")
print(p_a, ",", p_b, ",", p_c)
```

# Ejercicio 16

Trabajo en equipo. En un grupo de 25 alumnos hay 13 chicas y 12 chicos. Se desean formar

equipos de trabajo para realizar una investigación:

a) ¿Cuántos equipos de cinco personas pueden formarse?

b) ¿Cuántos equipos pueden formarse si debe haber una chica?

c) ¿Cuántos equipos pueden formarse si debe haber dos chicas?

d) Determina las probabilidades correspondientes a los incisos anteriores.



```python
# Codigo 
# a
from math import factorial, comb

num_alumnos = 25
num_equipos = 5
t_equipos_a = comb(num_alumnos, num_equipos)
print("Equipo de cinco personas: ", t_equipos_a)

# b) 
n_chicas = 13
n_chicos = 12
t_equipos_b = comb(n_chicas, 1) * comb(n_chicos, num_equipos - 1)
print("Equipos con una chica: ", t_equipos_b)

# c) 
t_equipos_c = comb(n_chicas, 2) * comb(n_chicos, num_equipos - 2)
print("Equipos con dos chicas: ", t_equipos_c)

# d) Probabilidades
total = comb(num_alumnos, num_equipos)
po_a = t_equipos_a / total 
po_b = t_equipos_b / total 
po_c = t_equipos_c / total 

print("Probabilidad: ")
print(po_a, ",", po_b, ",", po_c)
```

# Ejercicio 17

Control de calidad. Se tiene una caja con 24 latas y se sabe que dos están contaminadas. Se van

a seleccionar tres latas al azar para someterlas a una prueba de control de calidad, es decir, para

medir los estándares de calidad de la empresa.

a) ¿Cuántas combinaciones de tres latas pueden hacerse?

b) ¿Cuál es la probabilidad de que se seleccione una lata contaminada para la prueba?

c) ¿Cuál es la probabilidad de que se seleccione al menos una lata contaminada para la prueba?

d) ¿Y la probabilidad de que no se elijan latas contaminadas para la prueba?

```python
# Codigo 
# a
from math import comb
t_combinaciones= comb(24, 3)
print("a) Número de combinaciones de tres latas que pueden hacerse:", t_combinaciones)

# b) 
comb_1_conta= comb(2, 1) * comb(22, 2)
p_1_conta = round(comb_1_conta / t_combinaciones, 3)
print("b) Probabilidad de seleccionar una lata contaminada:", p_1_conta)

# c) 
# 2 contaminadas
comb_2_conta = comb(2, 2) * comb(22, 1)

casos_favo1= comb_1_conta + comb_2_conta
p_total = round(casos_favo1 / t_combinaciones, 3)
print("c) Probabilidad de seleccionar al menos una lata contaminada:", p_total)
```
